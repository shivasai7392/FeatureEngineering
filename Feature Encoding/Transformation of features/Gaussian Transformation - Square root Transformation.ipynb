```python
import pandas as pd
import numpy as np
import scipy.stats as stat
import pylab
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("/home/shivasai7392/Desktop/DS/Datasets/Titanic/titanic_train.csv",usecols=["Fare","Age","Pclass","Survived"])
```

```python
## if u want to check whether a feature is normally for Gaussian distributed use 
### Q-Q plot

def plot_data(data,feature):
    plt.figure(figsize=(20,10))
    plt.subplot(1,2,1)
    data[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(data[feature],dist="norm",plot=pylab)
    plt.show()
    
```

```python
data.Age.fillna(data.Age.median(),inplace=True)
```

```python
plot_data(data,"Age")
```

```python
data["Age_sqareroot"]= np.sqrt(data.Age)
```
