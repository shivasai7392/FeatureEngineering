```python
data.Age.fillna(data.Age.median(),inplace=True)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats as stat
import pylab
```

```python
## if u want to check whether a feature is normally for Gaussian distributed use 
### Q-Q plot

def plot_data(data,feature):
    plt.figure(figsize=(20,10))
    plt.subplot(1,2,1)
    data[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(data[feature],dist="norm",plot=pylab)
    plt.show()
    
```

```python
plot_data(data,"Age")
```

```python
import numpy as np
```

```python
data["Age_log"] = np.log(data.Age)
```
